```python
import pandas as pd
```

```python
insurance = pd.read_csv('insurance.csv', sep=',')
```

```python
insurance.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
insurance_train, insurance_test = train_test_split(insurance, test_size=0.3, 
                                                 stratify=insurance['smoker'],
                                                random_state=42)

print(insurance_train.shape, insurance_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

features = ['children']
dt_classification = DecisionTreeClassifier(max_depth = 3) # Increase max_depth to see effect in the plot
dt_classification.fit(insurance_train[features], insurance_train['smoker'])
```

```python
def calculate_accuracy(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (predictions == actuals).sum() / len(actuals)
```

```python
predictionsOnTrainSet = dt_classification.predict(insurance_train[features])
predictionsOnTestSet = dt_classification.predict(insurance_test[features])

accuracyTrain = calculate_accuracy(predictionsOnTrainSet, insurance_train['smoker'])
accuracyTest = calculate_accuracy(predictionsOnTestSet, insurance_test['smoker'])

print("Accuracy on training set " + str(accuracyTrain))
print("Accuracy on test set " + str(accuracyTest))
```

```python
from sklearn import tree

def plot_tree_classification(model, features, class_names):
    # generate plot data
    graph = tree.plot_tree(model,
                          feature_names=features,
                          class_names=class_names,
                          filled=True)
    
    # display in the notebook
    return graph
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,25))

plot_tree_classification(dt_classification, features, insurance['smoker'].unique())
```
