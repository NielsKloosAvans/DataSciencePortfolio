```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.close('all')
```

```python
games = pd.read_csv('Assignment 4.csv', sep=',')
```

```python
games.head()
```

```python
import seaborn as sb
from matplotlib import rcParams
```

```python
rcParams['figure.figsize'] = 12, 12
```

```python
sb.countplot(y=games['Platform'], order=games['Platform'].value_counts().iloc[:10].index)

plt.title('10 most popular platforms for games')
plt.xlabel('Amount of games on the platform')
plt.ylabel('Platform')

plt.show()
```

```python
sb.countplot(y=games['Genre'], order=games['Genre'].value_counts().iloc[:5].index)

plt.title('Top 5 most popular game genres')
plt.xlabel('Amount of games in this list')
plt.ylabel('Game genre')

plt.show()
```

```python
games_mean_global_sales = games['Global_Sales'].groupby(games['Year']).mean()
```

```python
games_mean_global_sales.plot()
plt.title('Mean global sales of games over the years')
plt.ylabel('Global sales')
plt.xticks(np.arange(1980, 2025, step=5))

plt.show()
```

```python
games_mean_eu_sales = games['EU_Sales'].groupby(games['Year']).mean()
```

```python
games_mean_eu_sales.plot()
plt.title('Mean Europa sales of games over the years')
plt.ylabel('Europa sales')
plt.xticks(np.arange(1980, 2025, step=5))

plt.show()
```
